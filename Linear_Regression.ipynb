```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('testdata.csv')
df.head()
```

```python
temperature = df.iloc[:, 1:2].values #Input 
preasure = df.iloc[:, 2].values #output
```

```python
linreg = LinearRegression()
linreg.fit(temperature, preasure) #fit the model
#predict the output
predicted = linreg.predict(temperature)
#plot the results
plt.scatter(temperature, preasure, color='red')
plt.scatter(temperature, predicted, color='blue')
plt.title('Temperature vs Pressure')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.legend(['Real Data', 'Predicted'])
plt.grid()
plt.show()
```

```python
#predict the output for new data
new_data = np.array([[30]])
predicted_new = linreg.predict(new_data)
print(f"Predicted pressure for temperature {new_data[0][0]}: {predicted_new[0]}")
```
